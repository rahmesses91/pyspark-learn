Section A - Cleaning & Normalization

```python
test_string = "   ID_00421 - john   doe   (NEW YORK)   "
test_string
```

1. strip()
Removes whitespace (or specified characters) from both ends of a string.

Time Complexity: 
O(n) without characters where n is the length of the string
O(n*m) with characters where m is the length of the characters

```python

stripped_whitespaces = test_string.strip()
stripped_whitespaces

stripped_parenthesis = test_string.strip().strip(')')
stripped_parenthesis
```

2. lstrip()
Removes whitespace (or specified characters) from the left side only.

```python

left_stripped_string = test_string.lstrip()
print(left_stripped_string)

left_stripped_string_of_a_character = test_string.strip().lstrip('I')
print(left_stripped_string_of_a_character)
```

3. rstrip()
Removes whitespace (or specified characters) from the right side only.
