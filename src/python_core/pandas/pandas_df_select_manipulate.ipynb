## 3. Selection & Indexing

**Why it matters:** Data engineers rarely need entire DataFrames. You'll constantly extract specific columns for transformations, filter rows based on conditions, and slice data for processing. Mastering selection is fundamental to every data pipeline.

```python
import pandas as pd

# Sample data for this section
orders = pd.DataFrame({
    'order_id': [1, 2, 3, 4, 5],
    'customer_id': ['C1', 'C2', 'C1', 'C3', 'C2'],
    'product': ['laptop', 'phone', 'mouse', 'laptop', 'tablet'],
    'amount': [1200, 800, 25, 1200, 500],
    'status': ['completed', 'pending', 'completed', 'cancelled', 'pending']
})
orders
```

### 3.1 Column Selection: Single Column
Access a single column using bracket notation `df['column_name']`. Returns a **Series**.

**Why important:** Extract specific fields for transformations, calculations, or passing to functions. Single column selection is the building block for all data manipulation.

```python
# Select single column - returns a Series
amounts = orders['amount']
print("Type:", type(amounts))
print("\nValues:\n", amounts)

# Alternative: dot notation (only works for column names without spaces/special chars)
# orders.amount  # Same result, but less flexible
```
